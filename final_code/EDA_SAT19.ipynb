```python
eda_satca19_df = pd.read_csv('/Users/edwardmendoza/Documents/GA/project_1/sat19_ca_df_cleaned_final')
```

```python
eda_satca19_df.head()
```

```python
eda_satca19_df.drop(columns = ['county_code', 'cdc_code', 'school_code'], inplace = True)
```

```python
eda_satca19_df.head()
```

```python
eda_satca19_df['total_highschoolers_passed'] = eda_satca19_df['total_12th_graders_passed'] + eda_satca19_df['total_11th_graders_passed']
```

```python
eda_satca19_df.head()
```

```python
eda_satca19_df.head()
```

```python
eda_satca19_df['total_eligible_students_enrolled'] = eda_satca19_df['enroll_12'] + eda_satca19_df['11th_graders_enrolled']
```

```python
eda_satca19_df['total_test_takers'] = eda_satca19_df['num_of_senior_test_takers'] + eda_satca19_df['11th_grade_test_takers']
```

```python
eda_satca19_df['pct_test_takers_to_enrolled'] = eda_satca19_df['total_test_takers'] / eda_satca19_df['total_eligible_students_enrolled']
```

```python
eda_satca19_df.head()
```

```python
bottom_10_counties = eda_satca19_df.groupby('county')['pct_test_takers_to_enrolled'].agg(['mean']).sort_values('mean', ascending = True).head(10)
```

```python
bottom_10_counties
```

```python
bottom_10_counties.plot(kind = 'bar', 
                        title = "Bottom 10 Counties of SAT Participation", 
                        figsize = (7,7),
                        xlabel = 'County Name',
                        ylabel = 'Percent of SAT Test Takers to Enrolled Students',
                        fontsize = 12,
                        legend = False);
```

```python
top_10_counties = eda_satca19_df.groupby('county')['pct_test_takers_to_enrolled'].agg(['mean']).sort_values('mean', ascending = False).head(10)
```

```python
top_10_counties.plot(kind = 'bar', 
                    title = "Top 10 Counties of ACT Participation", 
                    figsize = (7,7),
                    xlabel = 'County Name',
                    ylabel = 'Percent of ACT Test Takers to Enrolled Students',
                    fontsize = 12,
                    legend = False);
```
